```python
import os
import csv
```

# Train Hotdog

```python
with open('train_hot_dog.csv', 'w') as csvfile:
    output = csv.writer(csvfile)
    
    for category in os.listdir('/Users/erwinh/Desktop/Notebooks/NotebooksNew/HotdogData/train_hot_dog/'):
        for image in os.listdir(os.path.join('/Users/erwinh/Desktop/Notebooks/NotebooksNew/HotdogData/train_hot_dog/', category)):
            output.writerow(['gs://XXXXX/HotdogData/train_hot_dog/hot_dog/{}'.format(image),category])
            
    
    
    
```
